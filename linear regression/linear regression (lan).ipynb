```python
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import torch
```

```python
torch.manual_seed(100)
np.random.seed(100)
```

```python
x_train = np.linspace(-1,1,100)
x_t= torch.from_numpy(x_train)
x_t = torch.unsqueeze(x_t,dim=1)
y = x_train**2 + 0.2*np.random.rand(100)
y_t = torch.from_numpy(y)
print(x_t)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x_t,y_t)
```

```python
from torch.autograd import Variable

x_t =torch.unsqueeze(torch.linspace(-1,1,100),dim=1)
y_t = x_t**2 + 0.2*torch.rand(x_t.size())

x_t , y_t = Variable(x_t),Variable(y_t)


class net(nn.Module):
    def __init__(self,n_input,n_hidden,n_output):
        super(net,self).__init__()     #这个是固定套路
        self.hidden = nn.Linear(n_input,n_hidden)
        self.predict = nn.Linear(n_hidden,n_output)
        
    
    def forward(self,x):
        x= torch.nn.functional.relu(self.hidden(x))
        x= self.predict(x)
        return x
dnet = net(1,11,1)


optimizer = torch.optim.SGD(dnet.parameters(),lr = 0.2)
loss_F =torch.nn.MSELoss()

for i in range(60):
    prediction = dnet(x_t)
    
    loss =loss_F(prediction,y_t)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if (i + 1) % 5 == 0:
        print('Epoch: [{}/{}], Loss: {:.4f}'.format(i+1, 60, loss.item()))   
```

# 使用第二种方法观察：

```python
x_n = np.linspace(1,5,100)
y_n = x_n*3+0.4*np.random.rand(100)
b= torch.from_numpy(x_n)
x = torch.unsqueeze(b,dim=1)
y= torch.from_numpy(y_n)
model = nn.Linear(1,1)
model(x)
optimizer = torch.optim.SGD(model.parameters(),lr=0.02)
loss_F = nn.MSELoss()
plt.ion()
plt.show()

for epoch in range(70):
    prediction = model(x)
    loss = loss_F(prediction,y)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1)%5 == 0:
        plt.cla()
        plt.scatter(x_n,y_n)
        f= prediction.data.numpy()
        plt.plot(x_,f,'y',lw=5)
        plt.plause(0.1)
        print("epoch:[{}/{}],loss:{:.4f}".format(epoch+1,70,loss.item()))
    
```
